```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\acostaco\OneDrive - Vrije Universiteit Brussel\PhD\Secondments\Cambridge\experimental\EIT\figures\error localization\boxplot data.csv')
colors =["#E24A4E","#5289DA","#DE3D82","#0FB5AE", "#4248CA", "#F68511"]
sns.set_theme(style="ticks", palette=sns.color_palette(colors))
A=sns.boxplot(
    x='Area', y='Localization error (mm)', hue='Sample',
    data = df,
    notch=True,
    width=0.5,
    order=["Area1", "Area2", "Area3", "Area4"])
A.set(xlabel=None)
plt.savefig('areaerrors.svg')
data = df.groupby(["Area","Sample"])["Localization error (mm)"].describe()
data.to_csv('data.csv')
```

```python
median = df.groupby(["Area","Sample"])["Localization error (mm)"].median()
median.to_csv('median.csv')
```
