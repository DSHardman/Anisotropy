```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from statannotations.Annotator import Annotator

df = pd.read_csv(r'C:\Users\acostaco\OneDrive - Vrije Universiteit Brussel\PhD\Secondments\Cambridge\experimental\EIT\figures\error localization\boxplot data.csv')
```

```python
df1 = df[df['Area']=="Area1"]
```

```python
from itertools import combinations
Samples = ["No pattern", "Parallel lines", "Radial", "Concentric circles", "Full CB"]
pairs = list(combinations(Samples,r=2))
```

```python
ax=sns.boxplot(
    x='Sample', y='Localization error (mm)',
    data = df1,
    notch=True,
    width=0.5)
annotator = Annotator(ax, pairs, data=df1, x='Sample', y='Localization error (mm)')
annotator.configure(test="Mann-Whitney")
annotator.apply_and_annotate()
plt.savefig('Statistical1.svg')
```
